```python
diabetes=pd.read_csv('/content/diabetes.zip')
diabetes.head()
```

```python

target_var=diabetes['class']
target_var.head()

```

```python
features=diabetes.drop('class',axis=1)
features.head()
```

```python
for column in features.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(features[column], target_var, alpha=0.5)
    plt.xlabel(column)
    plt.ylabel('Target')
    plt.title(f'Relationship between {column} and Target')
    plt.show()
```

```python
one_values = ["Male", "Yes"]
zero_values = ["Female","No"]

for column in features.columns:
    features[column] = features[column].replace(to_replace=one_values, value=1)
    features[column] = features[column].replace(to_replace=zero_values, value=0)
features
```

```python
one_values = ["Positive"]
zero_values = ["Negative"]

target_var = target_var.replace(to_replace=one_values, value=1)
target_var= target_var.replace(to_replace=zero_values, value=0)
target_var.head()
```

```python
for column in features.columns:
    features[column] = pd.to_numeric(features[column], errors='coerce')


x_normalized = (features - features.min())/(features.max() - features.min())
x_normalized.head()
```
